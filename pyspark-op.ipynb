```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,desc
```

```python
spark = SparkSession.builder.appName("Op App").getOrCreate()
```

```python
data = spark.read.csv("/kaggle/input/op-data/operations_management.csv", header=True, inferSchema=True)
```

```python
data.show()
```

```python
data.printSchema()
```

```python
data.columns
```

```python
result = data.select(["industry","value"]).filter((col("value") > 200) & (col("industry") != "total")).orderBy(desc("value"))
```

```python
result.show()
```

```python
data.createOrReplaceTempView("data")
```

```python
spark.sql("""SELECT industry, value 
FROM data 
WHERE value > 200 
AND industry != "total"
order by value desc
""").show()
```
