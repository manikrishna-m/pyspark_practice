```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
spark = SparkSession.builder.appName("FirstApp").getOrCreate()
```

```python
myschema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("friends", IntegerType(), True)
])

# Read the CSV file into a DataFrame using the custom schema
data = spark.read.csv("/kaggle/input/fakefriends/fakefriends.csv", header=False, schema=myschema)
```

```python
data.printSchema()
```

```python
output = data.select("id","name","age","friends").where(data.age>30).withColumn("insert_ts",F.current_timestamp()).orderBy("id")
```

```python
output.count()
```

```python
output.createOrReplaceTempView("people")
```

```python
spark.sql("select name,age,friends,insert_ts from people").show()
```
